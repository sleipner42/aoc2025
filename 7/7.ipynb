```python
from functools import cache

with open("input.txt") as d:
    data = d.read().splitlines()

splits = set()
max_y = len(data)
start = None

for y, line in enumerate(data):
    for x, char in enumerate(line):
        if char == '^':
            splits.add((x, y))
        elif char == 'S':
            start = (x, y)
```

```python
@cache
def get_splits(current_pos: tuple[int, int]) -> set[tuple[int, int]]:
    if current_pos in splits:
        new_splits = {current_pos}
        new_splits.update(get_splits((current_pos[0] + 1, current_pos[1] + 1)))
        new_splits.update(get_splits((current_pos[0] - 1, current_pos[1] + 1)))
        return new_splits
    else:
        return get_splits((current_pos[0], current_pos[1] + 1)) if current_pos[1] < max_y else set()

len(get_splits(start))
```

```python
@cache
def count_possible_paths(current_pos: tuple[int, int]) -> int:
    if current_pos in splits:
        return count_possible_paths((current_pos[0] + 1, current_pos[1] + 1)) + count_possible_paths((current_pos[0] - 1, current_pos[1] + 1))
    else:
        return count_possible_paths((current_pos[0], current_pos[1] + 1)) if current_pos[1] < max_y else 1

count_possible_paths(start)
```
