```python
from functools import cache
import math
from typing import Optional

with open("input.txt") as d:
    data = d.read().splitlines()

points = [(int(x), int(y), int(z)) for row in data for x, y, z in [row.split(',')]]
number_of_points = len(points)


def distance_between_points(p1: tuple[int, int, int], p2: tuple[int, int, int]) -> int: 
    
    x_dist = (p1[0] - p2[0])**2
    y_dist = (p1[1] - p2[1])**2
    z_dist = (p1[2] - p2[2])**2
    
    return x_dist + y_dist + z_dist


distances = {}

for p1 in points:
    for p2 in points:
        if p1 != p2 and (p2, p1) not in distances:
            distances[(p1, p2)] = distance_between_points(p1, p2)

circuits: list[set[tuple[int, int, int]]] = []
n = 0
old_points = None

for points, _ in sorted(distances.items(), key=lambda x: x[1]):
    p1, p2 = points

    
    c1: Optional[int] = None
    c2: Optional[int] = None

    for idx, c in enumerate(circuits):
        if (p1 in c):
            c1 = idx
        if (p2 in c):
            c2 = idx

        if len(c) == number_of_points:
            print(old_points[0][0] * old_points[1][0])
            raise


    if (c1 is not None) and (c2 is not None) and (c1 != c2):
        circuits[c1].update(circuits[c2])
        circuits.pop(c2)

    elif c1 == c2 and (c1 is not None) and (c2 is not None):
        pass
    
    elif c1 is not None:
        circuits[c1].add(p2)

    elif c2 is not None:
        circuits[c2].add(p1)

    else:
        circuits.append(set([p1, p2]))

    n+=1
    old_points = points
    if n == 999:
        print(math.prod([len(a) for a in sorted(circuits, key=lambda x: len(x), reverse=True)[:3]]))
```
