```python
from functools import cache
from tqdm import tqdm

with open("input.txt") as d:
    data = d.read().splitlines()

data = {v.split(':')[0] : [k for k in v.split(': ')[-1].split()] for v in data}
```

```python
@cache
def ways_to_out(current_pos: str) -> int:

    if current_pos == 'out':
        return 1
    
    else:
        next_steps = data[current_pos]
        return sum([ways_to_out(n) for n in next_steps])

ways_to_out('you')
```

```python
@cache
def ways_to_out_2(current_pos: str, dac: bool, fft: bool) -> int:

    dac_visited = current_pos == 'dac' or dac
    fft_visited = current_pos == 'fft' or fft

    if current_pos == 'out':
        return 1 if (dac and fft) else 0
    
    else:
        next_steps = data[current_pos]
        return sum([ways_to_out_2(n, dac = dac_visited, fft = fft_visited) for n in next_steps])

ways_to_out_2('svr', False, False)
```
