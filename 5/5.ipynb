```python
with open("input.txt") as d:

    data = d.read()


ranges, ings = data.split("\n\n")
ranges = [range(int(r.split("-")[0]), int(r.split("-")[1]) + 1) for r in ranges.splitlines()]

nr_of_fresh = 0
for ing in ings.splitlines():
    
    for r in ranges:
        if int(ing) in r:
            nr_of_fresh += 1
            break

nr_of_fresh
```

```python
sorted_range = sorted(ranges, key=lambda x: x[0])
idx = 0

while idx < len(sorted_range) - 1:

    first_range = sorted_range[idx]
    second_range = sorted_range[idx + 1]

    if first_range[-1] >= second_range[0]:
        sorted_range.remove(first_range)
        sorted_range.remove(second_range)
        new_range = range(first_range[0], max(first_range[-1], second_range[-1]) + 1)
        sorted_range.insert(idx, new_range)
    else:
        idx += 1

sum([len(r) for r in sorted_range])
```
