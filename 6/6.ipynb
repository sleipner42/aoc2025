```python
import math

with open("input.txt") as d:

    data = d.read()


nr_of_problems = len(data.splitlines()[0].split())
problems = [[] for _ in range(nr_of_problems)]
operators = [ops for ops in data.splitlines()[-1].split()]
cuts = []

for rows in data.splitlines()[:-1]:
    
    for n, nr in enumerate(rows.split()):
        problems[n].append(int(nr))

total = 0

for op, problem in zip(operators, problems):
    cuts.append(max([len(str(nr)) for nr in problem]))
    total += sum(problem) if op == '+' else math.prod(problem)

total
```

```python
problems_nr_2 = [[] for _ in range(nr_of_problems)]

for rows in data.splitlines()[:-1]:

    idx = 0
    for n, cut in enumerate(cuts):
        problems_nr_2[n].append(rows[idx:idx+cut])
        idx += cut + 1

total_2 = 0

for op, prob, cut in zip(operators, problems_nr_2, cuts):

    new_numbers = ["" for _ in range(cut)]

    for nr in prob:
        for n, c in enumerate(nr[::-1]):
            new_numbers[n] += c

    total_2 += sum([int(nr) for nr in new_numbers]) if op == '+' else math.prod([int(nr) for nr in new_numbers])

total_2
```
