```python
from functools import cache
from tqdm import tqdm

with open("input.txt") as d:
    data = d.read().splitlines()

points = [(int(x), int(y)) for row in data for x, y in [row.split(',')]]


largest_area = 0

for p1 in points:
    for p2 in points:
        if p1 == p2:
            continue

        area = abs(p1[0] - p2[0] + 1) * abs(p1[1] - p2[1] + 1)
        if area > largest_area:
            largest_area = area

print(largest_area)
```
