```python
with open("input.txt") as d:

    data = d.read().splitlines()

total_joltage = 0

for line in data:

    numbers = [int(n) for n in line]

    max_n = max(numbers[:len(numbers)-1])
    idx_of_max = numbers.index(max_n)

    next_large_number = max(numbers[idx_of_max+1:])

    joltage = str(max_n) + str(next_large_number)

    total_joltage += int(joltage)

total_joltage
```

```python
total_joltage = 0

for line in data:

    numbers = [int(n) for n in line]

    max_n = 0
    idx_of_max = -1
    next_large_number = 0

    joltage = ""
    for i in range(12,0,-1):
        max_n = max(numbers[idx_of_max+1:len(numbers)-i+1])
        idx_of_max = numbers[idx_of_max+1:].index(max_n) + idx_of_max + 1
        joltage += str(max_n)

    total_joltage += int(joltage)

total_joltage
```
