```python
import re

with open("input.txt") as d:

    data = d.read().replace('\n', '').split(',')


def is_invalid(n):

    first_half = n[:len(n)//2]
    second_half = n[len(n)//2:]

    if first_half == second_half:
        return True

    return False
    

invalid_ids = []

for i in data:
    n1, n2 = i.split('-')
    for n in range(int(n1), int(n2)+1):
        if is_invalid(str(n)):
            invalid_ids.append(n)



sum(invalid_ids)
```

```python
def is_invalid(n):

    for j in range(0, len(n)//2 + 1):

        substring = n[0:j]
        matches = re.findall(substring, n)

        if len(matches) > 1 and len(matches) * len(substring) == len(n):
            return True

    return False
    

invalid_ids = []

for i in data:
    n1, n2 = i.split('-')
    for n in range(int(n1), int(n2)+1):
        if is_invalid(str(n)):
            invalid_ids.append(n)



sum(invalid_ids)
```
