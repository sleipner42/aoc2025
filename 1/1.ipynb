```python
with open("input.txt") as d:

    data = d.read().splitlines()


nr_is_zero = 0
current_nr = 50
max_nr = 100

for i in data:
    d = i[0]
    steps = int(i[1:])
    
    if d == "L":
        new_nr = current_nr - steps
    else:
        new_nr = current_nr + steps

    new_nr = new_nr % max_nr

    if new_nr == 0:
        nr_is_zero += 1

    current_nr = new_nr

print(nr_is_zero)
        
```
